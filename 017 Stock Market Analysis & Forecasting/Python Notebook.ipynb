# **Price Forecasting**

## **Get Data**

```python
AAPL_Stock_Price_3Y = yf.download('AAPL',
                                   start = Today_Time - timedelta(days=365*3),
                                   end=Today_Time,progress=False)

display(AAPL_Stock_Price_3Y.head())

plt.figure(figsize=(9,4))
plt.plot(AAPL_Stock_Price_3Y['Adj Close'])
plt.title('AAPL Stock Price')
plt.xlabel('Date',fontsize=16,labelpad=10)
plt.ylabel('Adj Close Price (USD)',fontsize=16,labelpad=5)
plt.show()
```

## **Select Data**

```python
AAPL_Price = AAPL_Stock_Price_3Y['Adj Close'].values
AAPL_Price = AAPL_Price.reshape(-1,1)

print(AAPL_Price.shape)
```

```python
print(AAPL_Price.shape[0])
AAPL_Price[:20]
```

```python
AAPL_Price_Train_Len = int(np.ceil(AAPL_Price.shape[0] * 0.90))
print('Training Data Size: ',AAPL_Price_Train_Len)
```

## **Split Data - Train Test**

```python
AAPL_Price_Train = AAPL_Price[0:AAPL_Price_Train_Len,:]
AAPL_Price_Test = AAPL_Price[AAPL_Price_Train_Len - 60:,:]

print(AAPL_Price_Train.shape)
print(AAPL_Price_Test.shape)
```

```python
# Convert 1D Time Series Data Into 2D Data 

def create_dataset(dataset,time_window):
    ''' 
    Convert 1D Time Series Data Into 2D Data
    Uses previous time series data to be used as predictors for the data at next time step.
    ----------------------------------------------------------
    Input:
    dataset: 1D Time Series Data [Array]
    time_window: Time Window Size [Int]
    '''
    import numpy as np
    data_X,data_Y = [],[]

    for i in range(0,len(dataset)-time_window-1):
        a = dataset[i:(i+time_window),0]
        data_X.append(a)
        data_Y.append(dataset[i + time_window,0])
    
    return np.array(data_X),np.array(data_Y)
```

```python
Train_X,Train_y = create_dataset(dataset=AAPL_Price_Train,time_window=60)
Test_X,Test_y = create_dataset(dataset=AAPL_Price_Test,time_window=60)

print(Train_X.shape, Train_y.shape)
print(Test_X.shape, Test_y.shape)
```

```python
display(Train_X[:10])
print('\n')
display(Test_X[:10])
```

## **Data Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
Scaler01 = MinMaxScaler()

Train_X = Scaler01.fit_transform(Train_X)
Test_X = Scaler01.fit_transform(Test_X)

display(Train_X[:10])
print('\n')
display(Test_X[:10])
```

## **Reshape Array**

```python
Train_X = Train_X.reshape(Train_X.shape[0],Train_X.shape[1],1)
Test_X = Test_X.reshape(Test_X.shape[0],Test_X.shape[1],1)

print(Train_X.shape, Train_y.shape)
print(Test_X.shape, Test_y.shape)
```

## **Create Model**

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
from sklearn.metrics import root_mean_squared_error

Model01 = Sequential()
Model01.add(LSTM(units=128,return_sequences=True,input_shape=(Train_X.shape[1],1)))
Model01.add(LSTM(units=64,return_sequences=False))
# Model01.add(LSTM(units=32,return_sequences=False))
# Model01.add(LSTM(units=16,return_sequences=False))
Model01.add(Dense(units=8))
Model01.add(Dense(units=1))

Model01.compile(optimizer='adam',loss='mean_squared_error',metrics=['accuracy'])
Model01.fit(Train_X,Train_y,epochs=20,batch_size=30)
```

```python
Test_y_Pred = Model01.predict(Test_X)
# Test_y_Pred = Scaler01.inverse_transform(Test_y_Pred)
Test_y_Pred = Test_y_Pred.reshape(-1)

display(Test_y[:10])
display(Test_y_Pred[:10])

Model01_RMSE = root_mean_squared_error(Test_y_Pred,Test_y)
print('Model01_RMSE:', Model01_RMSE)
```
